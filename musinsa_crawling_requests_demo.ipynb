```python
products = soup.select('a.img-block')
len(products)
```

```python
product_links = []

# while True:
for product in products:
    product_links.append(product['href'].replace('//', 'https://'))

product_links
```

```python
response = requests.get(product_links[0], headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
soup
```

```python
want = soup.find_all('script', {'type':'text/javascript'})[15]
want = want.string

pattern = re.compile(r'window\.__MSS__\.product\.state = ({.*?});\s*$', re.DOTALL)
match = pattern.search(want)
want2 = match.group(1)
json.loads(want2)
```

```python
test = []
pattern = re.compile(r'window\.__MSS__\.product\.state = ({.*?});\s*$', re.DOTALL)

for product_link in product_links:
    response = requests.get(product_link, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    want = soup.find_all('script', {'type':'text/javascript'})[15]
    want = want.string

    match = pattern.search(want)
    want2 = match.group(1)
    test.append(json.loads(want2))
    print(len(test))
```
